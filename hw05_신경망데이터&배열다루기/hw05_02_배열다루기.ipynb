# 배열 다루기 루틴

```python
import numpy as np
```

Q1. x가 10x10x3의 다차원 배열일때, x의 두번째 차원이 150인 2차원 배열이 되도록 reshape하세요.

```python
x = np.ones([10, 10, 3])

x_reshape = x.reshape(2, 150)
print(x_reshape.shape)
```

Q2. x가 [[1, 2, 3], [4, 5, 6]]일 때, [1 2 3 4 5 6]로 변환하세요.

```python
x = np.array([[1, 2, 3], [4, 5, 6]])
```

```python
x_flatten = x.flatten()
print(x_flatten)
```

```python
x_ravel = x.ravel()
print(x_ravel)
```

Q3. x가 [[1, 2, 3], [4, 5, 6]]일 때, flatten한 후 5번째 원소를 가져오세요.

```python
x = np.array([[1, 2, 3], [4, 5, 6]])
out = x.flatten()

print(out[4])
```

Q4.

x= <br/>
[[ 1 2 3]<br/>
[ 4 5 6].<br/><br/>
y = <br/>
[[ 7 8 9]<br/>
[10 11 12]].<br/>


x와 y를 연결해서
<br/>[[1, 2, 3, 7, 8, 9], <br/>[4, 5, 6, 10, 11, 12]]
를 만드세요


```python
x = np.array([[1, 2, 3], [4, 5, 6]])
y = np.array([[7, 8, 9], [10, 11, 12]])

z = np.hstack((x, y))
print(z)
```

Q5.

[[ 1 2 3]<br/>

[[ 7 8 9]<br/>
[10 11 12]].<br/>


x와 y를 연결해서
<br/>[[ 1  2  3]<br/>
 [ 4  5  6]<br/>
 [ 7  8  9]<br/>
 [10 11 12]]
를 만드세요

```python
x = np.array([[1, 2, 3], [4, 5, 6]])
y = np.array([[7, 8, 9], [10, 11, 12]])

z = np.vstack((x, y))
print(z)
```

Q6. x가 [0, 1, 2]일 때, [0, 0, 1, 1, 2, 2]를 생성하세요.

```python
x = np.array([0, 1, 2])

x_repeat = np.repeat(x, 2)

print(x_repeat)
```

Q7. x가 [0, 0, 0, 1, 2, 3, 0, 2, 1, 0]일 때,<br/>
앞, 뒤의 0을 제거하세요.

```python
x = np.array((0, 0, 0, 1, 2, 3, 0, 2, 1, 0))

x_trim = np.trim_zeros(x, 'fb')
print(x_trim)
```
